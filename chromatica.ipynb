# üé® Chromatica ‚Äî Mood-Driven Color Palette Generator

Welcome to **Chromatica**! This project takes a mood or concept word and generates a visually appealing color palette matching that theme. 

### Core Features
1. **Mood Interpretation**: Maps words to HSV (Hue, Saturation, Value) profiles.
2. **Palette Generation**: Creates 5 harmonious colors in HEX format.
3. **Visualization**: Displays the palette as clean horizontal blocks with labels.
4. **Reproducibility**: Supports optional random seeds for consistent results.

## üß¨ Reproducibility Test
Running the same mood with the same seed will always produce the same colors.

```python
mood = "energetic"
seed_value = 123

p1 = engine.generate_palette(mood, seed=seed_value)
p2 = engine.generate_palette(mood, seed=seed_value)

print(f"Palette 1: {p1}")
print(f"Palette 2: {p2}")

if p1 == p2:
    print("‚úÖ Success: Palettes are identical!")
else:
    print("‚ùå Error: Palettes differ.")
```
